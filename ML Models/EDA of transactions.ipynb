The dataset contains 2,000 rows and 17 columns related to blockchain transactions. Key columns include:

- **hash**: Transaction hash (unique identifier)
- **nonce**: Number of transactions sent from the sender's address
- **transaction_index**: Position of the transaction in the block
- **from_address**: Sender's address
- **to_address**: Recipient's address (1 missing value)
- **ether_transferred**: Amount of Ether transferred in the transaction
- **gas**: Gas provided for the transaction
- **gas_price**: Price of each gas unit
- **input**: Additional data sent with the transaction
- **receipt_cumulative_gas_used**: Total gas used when the transaction was executed
- **receipt_gas_used**: Gas used for the transaction
- **receipt_contract_address**: Contract address created by the transaction (mostly null)
- **receipt_status**: Transaction status (1 for success)
- **block_timestamp**: Time when the block was created
- **block_hash**: Hash of the block containing the transaction
- **block_number**: Block number in which the transaction was included

```python
import pandas as pd
import numpy as np

# Load dataset
#df = pd.read_csv('path_to_your_file.csv')

# 1. Time-based Features
df['block_timestamp'] = pd.to_datetime(df['block_timestamp'])
df['year'] = df['block_timestamp'].dt.year
df['month'] = df['block_timestamp'].dt.month
df['day'] = df['block_timestamp'].dt.day
df['hour'] = df['block_timestamp'].dt.hour

# 2. Gas Efficiency (Ether Transferred per Gas Used)
df['gas_efficiency'] = df['ether_transferred'] / df['receipt_gas_used']
df['gas_efficiency'].replace([np.inf, -np.inf], np.nan, inplace=True)

# 3. Transaction Size Feature
df['transaction_size'] = df['ether_transferred'] * df['gas']

# 4. Interaction Frequency
df['num_transactions_from_address'] = df.groupby('from_address')['from_address'].transform('count')
df['num_transactions_to_address'] = df.groupby('to_address')['to_address'].transform('count')

# 5. High Gas Indicator (flagging transactions with gas above 95th percentile)
high_gas_threshold = df['gas'].quantile(0.95)
df['high_gas_flag'] = np.where(df['gas'] > high_gas_threshold, 1, 0)

# 6. Outlier Detection based on Ether Transferred
ether_threshold = df['ether_transferred'].quantile(0.95)
df['high_ether_flag'] = np.where(df['ether_transferred'] > ether_threshold, 1, 0)

# Inspect the new features
df.head()
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Load the dataset
#df = pd.read_csv('path_to_your_file.csv')

# Filter necessary columns
df_graph = df[['from_address', 'to_address', 'gas', 'gas_price', 'ether_transferred']].dropna()

# Initialize a directed graph
G = nx.DiGraph()

# Add edges (transactions between addresses) with gas as the weight
for idx, row in df_graph.iterrows():
    from_node = row['from_address']
    to_node = row['to_address']
    
    # Use 'gas' as the edge weight (could also use other features like 'gas_price', 'ether_transferred')
    G.add_edge(from_node, to_node, weight=row['gas'])

# Plot the graph
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G, k=0.1)  # Layout for visualization
nx.draw(G, pos, with_labels=False, node_size=60, node_color="blue", alpha=0.6, edge_color="gray")
plt.title('Blockchain Transaction Graph')
plt.show()
```

```python
import torch
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torch_geometric.data import Data
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Assuming df_graph is already filtered and contains 'from_address', 'to_address', 'gas', and 'gas_price'

# Convert all addresses to strings (this will prevent mixing types)
df_graph['from_address'] = df_graph['from_address'].astype(str)
df_graph['to_address'] = df_graph['to_address'].astype(str)

# Combine all unique addresses from both columns
all_addresses = pd.concat([df_graph['from_address'], df_graph['to_address']]).unique()

# Encode node addresses as numerical IDs
le = LabelEncoder()
le.fit(all_addresses)

# Transform 'from_address' and 'to_address'
df_graph['from_address'] = df_graph['from_address'].apply(lambda x: le.transform([x])[0] if x in le.classes_ else -1)
df_graph['to_address'] = df_graph['to_address'].apply(lambda x: le.transform([x])[0] if x in le.classes_ else -1)

# Ensure no addresses are left as -1 (unseen addresses), by filtering them out
df_graph = df_graph[(df_graph['from_address'] != -1) & (df_graph['to_address'] != -1)]

# Create edge index and edge attributes (weights)
edge_index = torch.tensor([df_graph['from_address'].values, df_graph['to_address'].values], dtype=torch.long)
edge_weight = torch.tensor(df_graph['gas'].values, dtype=torch.float)

# Node features (initially random, can be improved later)
num_nodes = len(le.classes_)
node_features = torch.randn(num_nodes, 16)  # Initialize with random features

# Create a PyTorch Geometric data object
data = Data(x=node_features, edge_index=edge_index, edge_attr=edge_weight)

# Define a simple GNN model (Graph Convolutional Network)
class GNN(torch.nn.Module):
    def __init__(self):
        super(GNN, self).__init__()
        self.conv1 = GCNConv(16, 32)
        self.conv2 = GCNConv(32, 32)
        self.conv3 = GCNConv(32, 16)

    def forward(self, data):
        x, edge_index, edge_weight = data.x, data.edge_index, data.edge_attr
        x = self.conv1(x, edge_index, edge_weight)
        x = F.relu(x)
        x = self.conv2(x, edge_index, edge_weight)
        x = F.relu(x)
        x = self.conv3(x, edge_index, edge_weight)
        return x

# Initialize the model and optimizer
model = GNN()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

# Training loop (simplified)
for epoch in range(100):
    model.train()
    optimizer.zero_grad()
    out = model(data)
    loss = F.mse_loss(out[data.edge_index[0]], out[data.edge_index[1]])  # Minimize difference in embeddings
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        print(f'Epoch {epoch}, Loss: {loss.item()}')

# The model will learn node embeddings which can be used to optimize routes.
```
